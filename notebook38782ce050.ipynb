```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/project/Sample_Submission.csv")
df.head()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Check missing values
df.isnull().sum()

# Remove missing rows
df = df.dropna()
```

```python
print(df.columns)
```

```python
df = df.rename(columns={"Time_taken (min)": "delivery_time"})
df["delivery_time"] = df["delivery_time"].astype(int)
df["ID"] = df["ID"].astype(str)
df["late_delivery"] = df["late_delivery"].astype(int)
```

```python
df.head()
```

```python
df["late_delivery"] = df["delivery_time"] > 45
```

```python
df = df.dropna()
```

```python
df.to_csv("clean_orders.csv", index=False)
```
